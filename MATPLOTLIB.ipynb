```python
import matplotlib.pyplot as plt
import numpy as np

xpoints = np.array([0, 6,3,4,2])
ypoints = np.array([5, 15,9,4,3])

plt.plot(xpoints, ypoints)
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.array(["clad", "it", "cse", "phy","single variable","several varable"])
y = np.array([3, 8, 5, 10,7,7])

plt.bar(x,y)
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

y = np.array([35, 25, 25, 15])

plt.pie(y)
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

y = np.array([35, 25, 25, 15])
mylabels = ["c language", "phy", "maths", "clad"]

plt.pie(y, labels = mylabels)
plt.show()
```
